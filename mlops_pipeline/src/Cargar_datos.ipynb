## 1. Importaciones

```python
# Importación de la librería principal para manipulación de datos tabulares.
# pandas es el estándar de facto en el ecosistema Python para trabajar con DataFrames.
import pandas as pd

# Se importa también openpyxl para asegurar que pandas pueda leer archivos .xlsx.
# pandas usa openpyxl internamente como motor (engine) cuando lee Excel modernos.
# La importación explícita no es estrictamente necesaria, pero deja constancia
# de la dependencia del proyecto.
import openpyxl  # noqa: F401  (importado para documentar dependencia)

# Verificación rápida de versiones instaladas.
print(f'pandas:   {pd.__version__}')
print(f'openpyxl: {openpyxl.__version__}')
```

## 2. Definición de la ruta del dataset

El dataset `Base_de_datos.xlsx` se encuentra en la **raíz del repositorio**, según la estructura definida por el Proyecto Integrador.

Este notebook vive en `mlops_pipeline/src/`, por lo que la ruta relativa al dataset es subir dos niveles (`../../`) y luego acceder al archivo.

```python
# Ruta relativa al dataset desde la ubicación del notebook.
# Estructura del repositorio:
#   repo/
#     Base_de_datos.xlsx        <-- archivo a cargar
#     mlops_pipeline/
#       src/
#         Cargar_datos.ipynb    <-- ubicación de este notebook
RUTA_DATASET = '../../Base_de_datos.xlsx'

print(f'Ruta configurada para el dataset: {RUTA_DATASET}')
```

## 3. Carga del dataset

Se utiliza `pd.read_excel()` para leer el archivo. Como el archivo solo contiene una hoja (`Hoja1`), no es necesario especificar el parámetro `sheet_name`; pandas tomará la primera por defecto.

**Nota técnica:** la lectura de archivos Excel es notablemente más lenta que la de archivos CSV. Para datasets muy grandes en producción se recomienda convertir a CSV o formatos binarios optimizados (Parquet, Feather). Para este proyecto académico, el volumen (~10K filas) hace que el formato Excel sea aceptable.

```python
# Carga del dataset desde el archivo Excel.
# El método read_excel devuelve directamente un DataFrame.
df = pd.read_excel(RUTA_DATASET)

# Confirmación de carga exitosa mediante las dimensiones del DataFrame resultante.
print(f'Dataset cargado correctamente.')
print(f'Dimensiones (filas, columnas): {df.shape}')
```

## 4. Verificación mínima de la carga

Se realiza una única verificación mínima para confirmar que la carga fue exitosa: mostrar las primeras filas del DataFrame.

Esta es una **verificación de sanidad**, no un análisis exploratorio. El análisis exploratorio (tipos de datos, nulos, distribuciones, etc.) se realiza en su totalidad en `comprension_eda.ipynb`.

```python
# Visualización de las primeras 5 filas como verificación de carga.
df.head()
```
